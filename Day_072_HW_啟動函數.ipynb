# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return x * (x > 0)

def dReLU(x):
    return 1 * (x > 0)

x = plt.linspace(-10,10,100)
f1, = plt.plot(x, ReLU(x), "r")
f2, = plt.plot(x, dReLU(x), "b")
plt.grid()
plt.title("ReLU Function")
plt.legend([f1,f2], ["ReLU","Derivative of ReLU"], loc="upper left")
plt.show()
```

## Leaky ReLU

```python
import matplotlib.pyplot as plt
import numpy as np

def leaky_relu(x, alpha=0.01):
    return x * (x > 0) + alpha * x * (x <= 0)

def dleaky_relu(x, alpha=0.01):
    return 1 * (x > 0) + alpha * 1 * (x <= 0)

x = np.linspace(-10, 10, 100)
leaky_relu_values = leaky_relu(x)
dleaky_relu_values = dleaky_relu(x)

# 繪製圖表
plt.plot(x, leaky_relu_values, label='Leaky ReLU', color='r')
plt.plot(x, dleaky_relu_values, label='Derivative of Leaky ReLU', color='b')
plt.grid()
plt.title('Leaky ReLU Function')
plt.legend()
plt.show()
```
